Pegar a tabela base, remover as colunas desnecessárias, formatar e converter dados de last price e %, dps popular outra tabela com last_price com valores corretos

```python
import pandas as pd
import requests
```

```python
def converte_last_price(crypto_final):
    last_price = []
    for value in crypto_final['Last price']:
        if (value.count(',')):
            value = value.replace(',', '')

        value = value.removeprefix("$ ")
        
        value = float(value)
            
        last_price.append(value)
    
    return last_price
```

```python
def converte_porcentagem(crypto_final):
    porcentagem = []
    for value in crypto_final['%']:

        value = value.removesuffix('%')
        value = value.removeprefix("+")

        value = float(value)

        porcentagem.append(value)
    
    return porcentagem
```

```python
url = 'https://www.worldcoinindex.com/'
crypto_url = requests.get(url)
crypto_url
```

```python
crypto_data = pd.read_html(crypto_url.text)
crypto_data
```

```python
len(crypto_data), type(crypto_data)
crypto_data = crypto_data[0]
```

```python
crypto_final = crypto_data.iloc[:,2:6]
crypto_final
```

```python
filtro = crypto_final['Last price'] != "googletag.cmd.push(function() { googletag.display('div-gpt-ad-1571049370698-0'); });"

crypto_final = crypto_final[filtro]

filtro = crypto_final['Last price'] != "googletag.cmd.push(function() { googletag.display('div-gpt-ad-1571049390094-0'); });"

crypto_final = crypto_final[filtro]

crypto_final.head()
```

```python
last_price = converte_last_price(crypto_final)
porcentagem = converte_porcentagem(crypto_final)
```

```python
data = {'Name': crypto_final['Name'], "last_price": last_price, "%": porcentagem}

crypto_convertido = pd.DataFrame(data)
crypto_convertido
```

```python
dados_processados = pd.DataFrame({
    "Coluna": ["last_price", "%"], 
    "Média": [crypto_convertido['last_price'].mean(), crypto_convertido['%'].mean()],
    "Mediana": [crypto_convertido['last_price'].median(), crypto_convertido['%'].median()],
    "Moda": [crypto_convertido['last_price'].mode(), crypto_convertido['%'].mode()]
})

dados_processados.head()
```
